```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data.csv") 
# sonar data set 
# to determine object as rock or mine
# rock labeled as R and M for mine.
```

```python
data.head()
```

```python
X = data.drop('Target', axis=1)
y = data['Target']
```

## Implementing Trucated SVD

```python
from sklearn.decomposition import TruncatedSVD
```

```python
# 20 attributes is what i want from the set previous count of 60

svd = TruncatedSVD(n_components=20, n_iter=100, random_state=42)
X_new = svd.fit_transform(X)
```

```python
pd.DataFrame(X_new, columns=X.columns[:20])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
lr = LogisticRegression()
lr2 = LogisticRegression()
```

```python
lr.fit(X=X, y=y)
```

```python
y_pred = lr.predict(X)
```

### Before SVD

```python
print(classification_report(y, y_pred))
```

```python
lr2.fit(X_new, y)
```

```python
y_pred = lr2.predict(X_new)
```

### After SVD

```python
print(classification_report(y, y_pred))
```

## PCA example to remind

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=20)
X_new_pca = pca.fit_transform(X)
```

```python
lr_pca = LogisticRegression()
```

```python
lr_pca.fit(X_new_pca, y)
```

```python
y_pred = lr_pca.predict(X_new_pca)
```

```python
print(classification_report(y, y_pred))
```

```python
SIGMA = svd.singular_values_ # 20,20 (diagonal matrix)
```

```python
VT = svd.components_ # 20,60
```
